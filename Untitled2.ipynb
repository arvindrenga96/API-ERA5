```python
import os
import zipfile
import glob
import xarray as xr
import dask
import numpy as np

# Set the directory where your zip files are located
zip_directory = '/home/kumarv/renga/Documents/Fall_24/API-ERA5/Data'
# Set the directory where you want to extract the files
extract_directory = '/home/kumarv/renga/Documents/Fall_24/API-ERA5/Data_extracted'

# Create the extraction directory if it doesn't exist
os.makedirs(extract_directory, exist_ok=True)

# Step 1: Extract all zip files
zip_files = glob.glob(os.path.join(zip_directory, 'era5_data_*.zip'))
for zip_file in zip_files:
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(extract_directory)
    print(f"Extracted: {zip_file}")

# Step 2: Combine all .nc files into one NetCDF file using Dask, removing duplicate coordinates
nc_files = glob.glob(os.path.join(extract_directory, '*.nc'))

if nc_files:
    # Set Dask configuration to split large chunks
    with dask.config.set(**{'array.slicing.split_large_chunks': True}):
        # Open all .nc files as a multi-file dataset with Dask
        combined_ds = xr.open_mfdataset(nc_files, combine='by_coords', chunks={'time': 'auto', 'lat': 'auto', 'lon': 'auto'})
        
        # Remove duplicate longitude values
        _, unique_lon_indices = np.unique(combined_ds.lon, return_index=True)
        combined_ds = combined_ds.isel(lon=unique_lon_indices)

        # Remove duplicate latitude values (if any)
        _, unique_lat_indices = np.unique(combined_ds.lat, return_index=True)
        combined_ds = combined_ds.isel(lat=unique_lat_indices)

        print("Dataset after removing duplicates:")
        print(combined_ds)
        
        # Save the combined dataset to a new NetCDF file using Dask
        output_file = os.path.join(extract_directory, 'combined_unique_era5_data.nc')
        
        # Compute and save in parallel
        with dask.config.set(scheduler='threads', num_workers=4):  # Adjust num_workers as needed
            combined_ds.to_netcdf(output_file, compute=True)
        
        print(f"Combined NetCDF file with unique coordinates created: {output_file}")
else:
    print("No .nc files found in the extraction directory.")

# Optional: Remove individual .nc files after combining
for nc_file in nc_files:
    os.remove(nc_file)
    print(f"Removed: {nc_file}")

print("Process completed.")
```

```python
import dask
import dask.array as da
import numpy as np

def to_numpy(xarray_da):
    """
    Convert an xarray DataArray to a NumPy array using Dask for efficient processing.
    
    :param xarray_da: xarray DataArray
    :return: NumPy array
    """
    # Convert to Dask array if it's not already
    if isinstance(xarray_da.data, da.Array):
        dask_array = xarray_da.data
    else:
        dask_array = da.from_array(xarray_da.data, chunks='auto')
    
    # Compute the Dask array to NumPy array
    with dask.config.set(scheduler='threads', num_workers=4):  # Adjust num_workers as needed
        numpy_array = dask_array.compute()
    
    return numpy_array

# Example usage
try:
    Solar_Radiation_Flux_numpy = to_numpy(combined_ds['Solar_Radiation_Flux'])

    print("Shape of Solar_Radiation_Flux_numpy:", Solar_Radiation_Flux_numpy.shape)
    print("Data type of Solar_Radiation_Flux_numpy:", Solar_Radiation_Flux_numpy.dtype)
    print("First few values of Solar_Radiation_Flux_numpy:", Solar_Radiation_Flux_numpy.flatten()[:5])
except AttributeError as e:
    print(f"An error occurred: {e}")
    print("This might be because 'combined_ds' is not defined or doesn't have 'Solar_Radiation_Flux' as a variable.")
    print("Make sure you have run the previous cell to create 'combined_ds' before running this cell.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_mean_solar_radiation_pcolormesh(Solar_Radiation_Flux_numpy, combined_ds):
    # Calculate the mean of the first time step
    mean_solar_radiation = np.nanmean(Solar_Radiation_Flux_numpy[:, :, :], axis=0)

    # Get latitude and longitude values
    lats = combined_ds.lat.values
    lons = combined_ds.lon.values

    # Create a mesh grid for plotting
    lon, lat = np.meshgrid(lons, lats)

    # Create the plot
    fig, ax = plt.subplots(figsize=(15, 10))
    
    # Use pcolormesh for faster plotting
    im = ax.pcolormesh(lon, lat, mean_solar_radiation, cmap='viridis', shading='auto')
    
    # Add colorbar
    cbar = fig.colorbar(im, ax=ax, orientation='vertical', pad=0.02)
    cbar.set_label('Mean Solar Radiation Flux')

    # Set title and labels
    ax.set_title('Mean Solar Radiation Flux (First Time Step)')
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')

    # Add gridlines
    ax.grid(True, linestyle='--', alpha=0.6)

    # Set aspect ratio to equal for a non-distorted map
    ax.set_aspect('equal')

    # Optionally, set longitude and latitude limits
    ax.set_xlim(lons.min(), lons.max())
    ax.set_ylim(lats.min(), lats.max())

    plt.tight_layout()
    plt.show()

# Usage
try:
    plot_mean_solar_radiation_pcolormesh(Solar_Radiation_Flux_numpy, combined_ds)
except NameError:
    print("Error: Solar_Radiation_Flux_numpy or combined_ds is not defined.")
    print("Make sure you've run the previous cells to create these variables.")
except AttributeError as e:
    print(f"AttributeError: {e}")
    print("This might be because combined_ds doesn't have 'lat' or 'lon' as coordinates.")
    print("Check the structure of your combined_ds.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

def plot_detailed_solar_radiation_cartopy(Solar_Radiation_Flux_numpy, combined_ds):
    # Calculate the mean of all time steps
    mean_solar_radiation = np.nanmean(Solar_Radiation_Flux_numpy[:, :, :], axis=0)

    # Get latitude and longitude values
    lats = combined_ds.lat.values
    lons = combined_ds.lon.values

    # Create a figure with two subplots: global view and zoomed view
    fig = plt.figure(figsize=(20, 10))
    
    # Global view
    ax1 = fig.add_subplot(1, 2, 1, projection=ccrs.PlateCarree())
    ax1.set_global()
    ax1.add_feature(cfeature.COASTLINE)
    ax1.add_feature(cfeature.BORDERS)
    
    # Plot the data with more levels for finer detail
    im1 = ax1.contourf(lons, lats, mean_solar_radiation, 
                       transform=ccrs.PlateCarree(), 
                       cmap='viridis', levels=50)
    
    # Add colorbar
    cbar1 = fig.colorbar(im1, ax=ax1, orientation='horizontal', pad=0.08)
    cbar1.set_label('Mean Solar Radiation Flux')
    
    ax1.set_title('Global Mean Solar Radiation Flux')
    ax1.gridlines(draw_labels=True, linewidth=1, color='gray', alpha=0.5, linestyle='--')

    # Zoomed view (focusing on the problematic area)
    ax2 = fig.add_subplot(1, 2, 2, projection=ccrs.PlateCarree())
    
    # Set the extent to focus on the problematic area (adjust these values as needed)
    ax2.set_extent([-180, 180, 90, 60], crs=ccrs.PlateCarree())
    ax2.add_feature(cfeature.COASTLINE)
    ax2.add_feature(cfeature.BORDERS)
    
    # Plot the data with even more levels for finer detail
    im2 = ax2.contourf(lons, lats, mean_solar_radiation, 
                       transform=ccrs.PlateCarree(), 
                       cmap='viridis', levels=100)
    
    # Add colorbar
    cbar2 = fig.colorbar(im2, ax=ax2, orientation='horizontal', pad=0.08)
    cbar2.set_label('Mean Solar Radiation Flux')
    
    ax2.set_title('Zoomed View of Problematic Area')
    ax2.gridlines(draw_labels=True, linewidth=1, color='gray', alpha=0.5, linestyle='--')

    plt.tight_layout()
    plt.show()

    # Print statistics for the problematic area
    print("Statistics for the zoomed area:")
    mask = (lats >= -90) & (lats <= -60)
    problematic_data = mean_solar_radiation[mask, :]
    print(f"Min value: {np.nanmin(problematic_data)}")
    print(f"Max value: {np.nanmax(problematic_data)}")
    print(f"Mean value: {np.nanmean(problematic_data)}")
    print(f"Number of NaN values: {np.sum(np.isnan(problematic_data))}")

# Usage
try:
    plot_detailed_solar_radiation_cartopy(Solar_Radiation_Flux_numpy, combined_ds)
except NameError:
    print("Error: Solar_Radiation_Flux_numpy or combined_ds is not defined.")
    print("Make sure you've run the previous cells to create these variables.")
except AttributeError as e:
    print(f"AttributeError: {e}")
    print("This might be because combined_ds doesn't have 'lat' or 'lon' as coordinates.")
    print("Check the structure of your combined_ds.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    print("Make sure you have Cartopy installed: pip install cartopy")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def analyze_solar_radiation_data(Solar_Radiation_Flux_numpy, combined_ds):
    # 1. Print basic information
    print(f"Shape of Solar_Radiation_Flux_numpy: {Solar_Radiation_Flux_numpy.shape}")
    print(f"Latitude range: {combined_ds.lat.min().values} to {combined_ds.lat.max().values}")
    print(f"Longitude range: {combined_ds.lon.min().values} to {combined_ds.lon.max().values}")

    # 2. Analyze polar regions
    north_pole = Solar_Radiation_Flux_numpy[:, -1, :]  # Assuming last index is northernmost
    south_pole = Solar_Radiation_Flux_numpy[:, 0, :]   # Assuming first index is southernmost

    print("\nNorth Pole Statistics:")
    print(f"Mean: {np.mean(north_pole)}")
    print(f"Std Dev: {np.std(north_pole)}")
    print(f"Min: {np.min(north_pole)}")
    print(f"Max: {np.max(north_pole)}")

    print("\nSouth Pole Statistics:")
    print(f"Mean: {np.mean(south_pole)}")
    print(f"Std Dev: {np.std(south_pole)}")
    print(f"Min: {np.min(south_pole)}")
    print(f"Max: {np.max(south_pole)}")

    # 3. Plot histograms
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    ax1.hist(north_pole.flatten(), bins=50)
    ax1.set_title("North Pole Solar Radiation Distribution")
    ax1.set_xlabel("Solar Radiation Flux")
    ax1.set_ylabel("Frequency")

    ax2.hist(south_pole.flatten(), bins=50)
    ax2.set_title("South Pole Solar Radiation Distribution")
    ax2.set_xlabel("Solar Radiation Flux")
    ax2.set_ylabel("Frequency")

    plt.tight_layout()
    plt.show()

    # 4. Check for constant values along longitudes at poles
    print("\nChecking for constant values along longitudes at poles:")
    print(f"North Pole constant: {np.allclose(north_pole, north_pole[:, [0]] * np.ones_like(north_pole))}")
    print(f"South Pole constant: {np.allclose(south_pole, south_pole[:, [0]] * np.ones_like(south_pole))}")

    # 5. Plot a slice of data near the poles
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10))

    im1 = ax1.imshow(Solar_Radiation_Flux_numpy[0, -10:, :], aspect='auto', cmap='viridis')
    ax1.set_title("Last 10 Latitude Rows (Near North Pole)")
    plt.colorbar(im1, ax=ax1)

    im2 = ax2.imshow(Solar_Radiation_Flux_numpy[0, :10, :], aspect='auto', cmap='viridis')
    ax2.set_title("First 10 Latitude Rows (Near South Pole)")
    plt.colorbar(im2, ax=ax2)

    plt.tight_layout()
    plt.show()

# Usage
try:
    analyze_solar_radiation_data(Solar_Radiation_Flux_numpy, combined_ds)
except NameError:
    print("Error: Solar_Radiation_Flux_numpy or combined_ds is not defined.")
    print("Make sure you've run the previous cells to create these variables.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

def explore_specific_region(Solar_Radiation_Flux_numpy, combined_ds):
    # Define the region of interest
    lat_min, lat_max = -79.9, 80
    lon_min, lon_max = -179.9, 180

    # Get latitude and longitude values
    lats = combined_ds.lat.values
    lons = combined_ds.lon.values

    # Find indices corresponding to the region of interest
    lat_indices = np.where((lats >= lat_min) & (lats <= lat_max))[0]
    lon_indices = np.where((lons >= lon_min) & (lons <= lon_max))[0]

    # Extract the data for the region of interest
    region_data = Solar_Radiation_Flux_numpy[:, lat_indices[:, None], lon_indices]

    # Calculate statistics
    nan_count = np.isnan(region_data).sum()
    total_elements = region_data.size
    nan_percentage = (nan_count / total_elements) * 100

    print(f"Region Statistics (Lat: {lat_min} to {lat_max}, Lon: {lon_min} to {lon_max}):")
    print(f"Total elements: {total_elements}")
    print(f"Number of NaN values: {nan_count}")
    print(f"Percentage of NaN values: {nan_percentage:.2f}%")

    if nan_count < total_elements:
        print(f"Minimum non-NaN value: {np.nanmin(region_data)}")
        print(f"Maximum non-NaN value: {np.nanmax(region_data)}")
        print(f"Mean of non-NaN values: {np.nanmean(region_data):.2f}")
    
    # Visualize the region
    mean_solar_radiation = np.nanmean(region_data, axis=0)

    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

    # Set the extent to our region of interest
    ax.set_extent([lon_min, lon_max, lat_min, lat_max], crs=ccrs.PlateCarree())

    # Plot the data
    im = ax.pcolormesh(lons[lon_indices], lats[lat_indices], mean_solar_radiation, 
                       transform=ccrs.PlateCarree(), 
                       cmap='viridis')

    # Add colorbar
    cbar = fig.colorbar(im, ax=ax, orientation='vertical', pad=0.08)
    cbar.set_label('Mean Solar Radiation Flux')

    ax.set_title(f'Mean Solar Radiation Flux (Lat: {lat_min} to {lat_max}, Lon: {lon_min} to {lon_max})')
    ax.gridlines(draw_labels=True, linewidth=1, color='gray', alpha=0.5, linestyle='--')

    plt.show()

# Usage
try:
    explore_specific_region(Solar_Radiation_Flux_numpy, combined_ds)
except NameError:
    print("Error: Solar_Radiation_Flux_numpy or combined_ds is not defined.")
    print("Make sure you've run the previous cells to create these variables.")
except AttributeError as e:
    print(f"AttributeError: {e}")
    print("This might be because combined_ds doesn't have 'lat' or 'lon' as coordinates.")
    print("Check the structure of your combined_ds.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
